```python
"""The program below illustrates calculating the Pythogorean distance between three points."""
```

```python
# Initializing the variables x,y and z
x = 2 
y = 3
z = 4
```

```python
# Print the values of x,y and z
print(x)
print(y)
print(z)
```

```python
# Calculate the sum of squares of three numbers
p = x**2 + y**2 + z**2
```

```python
# Print the value of the sum of squares of three numbers
print(p)
```

```python
# Calculate the square root of p
s = p**(1/2)
```

```python
# Print the value of the square root of p which is the Pythogorean distance between the three points
print(s)
```
